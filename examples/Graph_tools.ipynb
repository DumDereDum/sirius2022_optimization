## Documentation and examples

* **make_circular_graph_matrix(N)** - makes circular graph with __N__ nodes
* **make_complete_graph_matrix(N)** - makes complete graph with __N__ nodes
* **make_random_graph_matrix(N, occupancy=0.0, vanish_rate = 0)** - makes complete graph with __N__ nodes and removes __occupancy__ percent of edges or __vanish_rate__ ages


* **vanish_random_edges_from_graph(G, occupancy=0.0, vanish_rate = 0)** - removes __occupancy__ percent of edges or __vanish_rate__ ages from graph __G__, which is given as nx.Graph
* **vanish_random_edges_from_matrix(M, occupancy=0.0, vanish_rate = 0):** - removes __occupancy__ percent of edges or __vanish_rate__ ages from graph __G__, which is given as matrix
* **vanish_random_edge_from_matrix(M)** - removes one random edge from graph __M__, which is given as matrix


* **add_random_edge_to_matrix(M)** - add random edge to graph __M__, which is given as matrix


* __fill_metropolis_weigts(M)__ - fills graph __M__, which is given as matrix, by *metropolis weights* rule


* __make_graph_img(M, num_of_lines = 5, fig_size=(5, 5))__ - makes image of __M__, which is given as matrix, with size of __fig_size__ and lines are colored on __num_of_lines__ colors

```python
M = make_circular_graph_matrix(10)
M = fill_metropolis_weigts(M)
make_graph_img(M)
```

```python
M = make_complete_graph_matrix(10)
M = fill_metropolis_weigts(M)
make_graph_img(M)
```

```python
M = make_random_graph_matrix(6, occupancy=0.4)
M = fill_metropolis_weigts(M)
make_graph_img(M, fig_size=(5, 5))
```

```python
M = vanish_random_edge_from_matrix(M)
make_graph_img(M, fig_size=(5, 5))
```
