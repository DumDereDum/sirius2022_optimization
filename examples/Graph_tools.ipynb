## Documentation and examples

* __make_cirvular_graph(N)__ - makes circular graph with __N__ nodes
* __make_complete_graph(N)__ - makes complete graph with __N__ nodes
* __make_random_graph(N, occupancy=0.5)__ - makes complete graph with __N__ nodes and removes __occupancy__ percent of edges
* __vanish_random_edge(M)__ - vanishes random edge from graph __M__, which is given as matrix
* __fill_metropolis_weigts(M)__ - fills graph __M__, which is given as matrix, by *metropolis weights* rule
* __make_graph_img(M, num_of_lines = 5, fig_size=(5, 5))__ - makes image of __M__, which is given as matrix, with size of __fig_size__ and lines are colored on __num_of_lines__ colors

```python
M = make_cirvular_graph(10)
M = fill_metropolis_weigts(M)
make_graph_img(M)
```

```python
M = make_complete_graph(10)
M = fill_metropolis_weigts(M)
make_graph_img(M)
```

```python
M = make_random_graph(6)
M = fill_metropolis_weigts(M)
make_graph_img(M, fig_size=(5, 5))
```

```python
M = vanish_random_edge(M)
make_graph_img(M, fig_size=(5, 5))
```
